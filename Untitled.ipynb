## Request Hour

What we can do is divide the hours into 5 categories
1)Early Morning (5-10).When people may be visting
2)Day time(10-5)
3)Evening(5-9)
4)Night time(9-12)
5)Late at night(12-5)

```python
def hour_to_category(hour):
    if hour>=0 and hour<5:
        return "Late at night"
    elif hour>=5 and hour<10:
        return "Early Morning"
    elif hour>=10 and hour<17:
        return "Day time"
    elif hour>=17 and hour<21:
        return "Evening"
    else:
        return "Night time"
    
        
    
```

```python
data['Day time']=data['Request_hour'].apply(hour_to_category)
data
```

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(x="Day time", data=data,hue="Status",)

for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))



plt.show()
bx = sns.countplot(x="Day time", data=data)

for p in bx.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

As we can most rush is found during early morning and evening.In case of Evening No cars available is a huge problem followed by 
Night time and Late at nights!.In case of early morning a lot of rides are getting cancelled by the drivers which in itself
is a problem!

# Bivariate Analysis

```python
#Creating a dataset with different format for a clear analysis
```

```python
databivariate=data.drop(['Request id','Driver id','Request timestamp','Drop timestamp'],axis=1)
databivariate.sample(10)
```

Turning Status Column into three different categorical columns using pd.get_dummies

```python
databivariate=pd.get_dummies(databivariate,columns=['Status'])
```

```python
databivariate.columns=['Pickup point', 'Request_year', 'Request_month', 'Request_day',
       'Request_hour', 'Day time', 'Cancelled',
       'No Cars Available', 'Completed']
```

```python
df_bi=databivariate.groupby(['Request_day','Pickup point','Day time']).agg({'Cancelled':'sum','No Cars Available':'sum',
                                                                    'Completed':'sum'})
df_bi=df_bi.sort_values(['Request_day','Pickup point','Day time'],ascending=False)
df_bi
```

```python
#Converting it into data frame to calculate further measures
df_bi=pd.DataFrame(df_bi)
df_bi['Total Requests']=df_bi['Cancelled']+df_bi['No Cars Available']+df_bi['Completed']
df_bi['Cancellation_percentage']=(df_bi['Cancelled']/(df_bi['Cancelled']+df_bi['No Cars Available']+df_bi['Completed']))*100
df_bi['No Cars Available percentage']=(df_bi['No Cars Available']/(df_bi['Cancelled']+df_bi['No Cars Available']+df_bi['Completed']))*100
df_bi['Completed percentage']=(df_bi['Completed']/(df_bi['Cancelled']+df_bi['No Cars Available']+df_bi['Completed']))*100

df_bi.sort_values(['Request_day','Pickup point','Completed percentage','Day time'],ascending=False)
```

As we can see,In case of City,Completed trip percentage is highest during evening and lowest at Early morning.Also,during Early mornings a significant amount of these requests are not completed due to cancellations by the driver.
In case of airport,this is quite the opposite.Completed trip percentage is highest during early mornings(even higher than City during evenings) but this number is very poor at evening time(even lower than City during mornings).Also,during early mornings a significant amount of these requests are not completed due to non-availabilty due to cars.
Now this is a pure supply and demand issue as demand is highest during Early morning in case of City.That's why many requests are left incompleted.
Visa versa,Demand is highest during evening time in case of Airports.That's why there is a huge non-availabilty issue during these time.Let's try to understand it better with help of some graphs.

```python
df_bi2=df_bi.reset_index()
```

```python
corr = df_bi2.corr()

# Create a heat map of the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```

As we can see the total requests have a .55 correlation with cancellation percentage.It means that cancellation percentage will increase if demand is increased.
Visa versa total requests have a -.48 correlation with completed percentage.It means that completed percentage will decrease if demand is increased.
